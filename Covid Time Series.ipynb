```python
country_df=df[df['Country/Region']=='India']
```

```python
country_df=country_df.drop(columns=["Province/State","Lat","Long",'Country/Region'],errors='ignore')
```

```python
country_df.head()
```

```python
country_ts = country_df.sum(axis=0)

# Step 5: Convert index to datetime (skip non-date columns)
country_ts.index = pd.to_datetime(country_ts.index, errors='coerce')

# Step 6: Drop invalid dates (e.g., NaT caused by non-date columns)
country_ts = country_ts.dropna()

# Step 7: Convert the values (not the index) to integers
country_ts = country_ts.astype(int)

# Step 8: Compute daily new cases
daily_cases = country_ts.diff().fillna(0).astype(int)

# Step 9: Create final DataFrame for modeling
df_daily = pd.DataFrame({
    "ds": daily_cases.index,   # Prophet-compatible date column
    "y": daily_cases.values    # New daily confirmed cases
})

# Optional: Preview
print(df_daily.head())
```

```python
print(df_daily.head())
```

```python
import matplotlib.pyplot as plt

# Plot raw daily cases
plt.figure(figsize=(12, 5))
plt.plot(df_daily['ds'], df_daily['y'], label='Daily Cases', color='skyblue')
plt.title(f'Daily New COVID-19 Cases in India')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Add a 7-day rolling average column
df_daily['7_day_avg'] = df_daily['y'].rolling(window=7).mean()

# Plot daily cases with 7-day average
plt.figure(figsize=(12, 5))
plt.plot(df_daily['ds'], df_daily['y'], label='Daily Cases', alpha=0.4, color='gray')
plt.plot(df_daily['ds'], df_daily['7_day_avg'], label='7-Day Moving Average', color='blue', linewidth=2)
plt.title(f'Daily COVID-19 Cases in India with 7-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define a threshold to detect spikes (e.g., top 1% of cases)
threshold = df_daily['y'].quantile(0.95)

# Mark spikes
spikes = df_daily[df_daily['y'] > threshold]

plt.figure(figsize=(12, 5))
plt.plot(df_daily['ds'], df_daily['y'], label='Daily Cases', color='gray', alpha=0.4)
plt.scatter(spikes['ds'], spikes['y'], color='red', label='Spikes', zorder=5)
plt.title(f'Spikes in Daily COVID-19 Cases in India')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Set split date — train until this date
split_date = '2022-01-31'

# Train data
df_train = df_daily[df_daily['ds'] <= split_date]

# Test data (actual future data to compare predictions)
df_test = df_daily[df_daily['ds'] > split_date]
```

```python
model = Prophet()
model.fit(df_train)
```

```python
# Create future dataframe for the same length as test set
future = model.make_future_dataframe(periods=len(df_test))

# Generate forecast
forecast = model.predict(future)
```

```python
# Extract predicted values for the test period
pred_test = forecast[['ds', 'yhat']][forecast['ds'].isin(df_test['ds'])]

# Merge actual test values
merged = df_test.merge(pred_test, on='ds')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(merged['y'], merged['yhat'])
rmse = np.sqrt(mean_squared_error(merged['y'], merged['yhat']))

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
```

```python
plt.figure(figsize=(12, 5))
plt.plot(merged['ds'], merged['y'], label='Actual', color='black')
plt.plot(merged['ds'], merged['yhat'], label='Forecast', color='blue')
plt.title(f'Actual vs Forecasted COVID-19 Cases India')
plt.xlabel("Date")
plt.ylabel("Daily Cases")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Create a DataFrame of events
lockdowns = pd.DataFrame({
    'holiday': 'lockdown',
    'ds': pd.to_datetime([
        '2020-03-25', '2020-04-15', '2021-04-15'  # Add real or simulated lockdown dates
    ]),
    'lower_window': 0,
    'upper_window': 5  # Number of days you think the effect lasts
})

# Step 2: Initialize Prophet with events
model = Prophet(holidays=lockdowns)
model.fit(df_train)

# Step 3: Forecast as usual
future = model.make_future_dataframe(periods=len(df_test))
forecast = model.predict(future)

# Step 4: Plot
model.plot(forecast)
plt.title("Forecast with Lockdowns Included")
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

```

```python
series = df_daily.set_index('ds')['y']
```

```python
split_date = '2022-01-31'
train = series[:split_date]
test = series[split_date:]
```

```python
model = ARIMA(train, order=(2, 1, 2))
model_fit = model.fit()
```

```python
forecast = model_fit.forecast(steps=len(test))
```

```python
mae = mean_absolute_error(test, forecast)
rmse = np.sqrt(mean_squared_error(test, forecast))
print(f'MAE: {mae:.2f}')
print(f'RMSE: {rmse:.2f}')
```

```python
plt.figure(figsize=(12, 5))
plt.plot(train.index, train, label='Training Data', color='black')
plt.plot(test.index, test, label='Actual', color='green')
plt.plot(test.index, forecast, label='ARIMA Forecast', color='orange')
plt.title('ARIMA Forecast vs Actual COVID-19 Daily Cases')
plt.xlabel('Date')
plt.ylabel('Daily Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
